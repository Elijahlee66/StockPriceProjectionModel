```python
# Create a candle chart
import pandas_datareader as web
from datetime import date
enddate = date.today().strftime("%Y-%m-%d")
import plotly.graph_objects as go
import pandas as pd
```

```python
# Load in stock data
df = web.DataReader('JNJ',data_source = 'yahoo', start = '2020-01-01',end = enddate)
# df = df.set_index(pd.DatetimeIndex)
```

```python
# Create a candlestick chart
figure = go.Figure(
    data = [
            go.Candlestick(
                x = df.index,
                low = df['Low'],
                high = df['High'],
                close = df['Close'],
                open = df['Open'],
                increasing_line_color = 'red',
                decreasing_line_color = 'green'
            )
    ]
)

#figure.update_layout(xaxis_rangeslider_visible = False)
figure.update_layout(
    title = 'Johnson & Johnson Stock Price',
    yaxis_title = 'Price ($)',
    xaxis_title = 'Date'
    



)
figure.show()
```
